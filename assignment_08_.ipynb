```python
import pandas as pd
```

```python
df1 = pd.read_csv("https://tinyurl.com/4p76n89r", delimiter=';')
df2 = pd.read_csv("https://tinyurl.com/2faje6k9", delimiter='\t')

print("DataFrame 1:")
display(df1.head())

print("\nDataFrame 2:")
display(df2.head())
```

```python
display(df1.head())
display(df1.info())
display(df1.describe())
```

```python
display(df2.head())
display(df2.info())
display(df2.describe())
```

```python
merged_df = pd.merge(df1, df2, how='inner', on='Date')
display(merged_df)
```

```python
region_summary = merged_df.groupby('Region')['Amount_x'].agg(['mean', 'sum'])
display(region_summary)
```

```python
merged_df = merged_df.drop_duplicates()
print("DataFrame after removing duplicates:")
display(merged_df)
```

```python
merged_df = merged_df.loc[:, ~merged_df.columns.str.contains('Unnamed')]
display(merged_df)
```
